In the space below, write code to train a logistic regression model on the training set.

```python
# YOUR CODE GOES HERE
def sigmoid(z):   
    return 1 / (1 + np.exp(-z))

def h(X, theta):
    return sigmoid(X @ theta)

def grad_j(X, y, y_pred):
    return X.T @ (y - y_pred) / X.shape[0]
    
def j(theta, X, y):    
    y_pred = h(X, theta)
    error = (-y * np.log(y_pred)) - ((1 - y) * np.log(1 - y_pred))
    cost = sum(error) / X.shape[0]
    grad = grad_j(X, y, y_pred)
    return cost[0], grad

def train(X, y, theta_initial, alpha, num_iters):
    theta = theta_initial
    j_history = []
    for i in range(num_iters):
        cost, grad = j(theta, X, y)
        theta = theta + alpha * grad
        j_history.append(cost)
    return theta, j_history


alpha = .005
num_iters = 10000
theta_initial = np.zeros((X_train.shape[1], 1))
theta_initial.shape
theta, j_history = train(X_train, y_train, theta_initial, alpha, num_iters)
print(theta)

plt.plot(j_history)
plt.xlabel("Iteration")
plt.ylabel("$J(\theta)$")
plt.title("Training cost over time with batch gradient descent (no normalization)")
plt.show()

plt.figure()
plt.scatter(X_train[:,1], X_train[:,2], c=y_train)
x1 = np.linspace(np.amin(X_train[:,1]), np.amax(X_train[:,1]), 50)
x2 = -(theta[0] + x1*theta[1])/theta[2]
plt.plot(x1, x2)
plt.title('Training dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
import cvxopt

def cvxopt_solve_qp(Q, c, A=None, B=None, E=None, d=None):

    Q_new = (Q + Q.T)/2
    args = [cvxopt.matrix(Q_new), cvxopt.matrix(c)]
    if A is not None:
        args.extend([cvxopt.matrix(A), cvxopt.matrix(B)])
        if E is not None:
            args.extend([cvxopt.matrix(E), cvxopt.matrix(d)])
            
    sol = cvxopt.solvers.qp(*args)
    
    if sol is not None and 'optimal' not in sol['status']:
        return None
#     print(sol)
    x = sol['x']
    
    return x

def gausskernel(X, sigma):
    m = X.shape[0];
    K = np.matrix(np.zeros([m,m]));
    for i in range(0,m):
        for j in range(0,m):
            K[i,j] = (X[i,:] - X[j,:]).reshape(1,-1) @ (X[i,:] - X[j,:]).reshape(-1,1) #||X-Z||^2 = (X-Z).T @(X-Z)
    K = np.exp(-K/(2*sigma*sigma))      
    return K;


def quadparams(X, y, K, c):
    Q = np.diag(np.array(y).squeeze()) @ K @ np.diag(np.array(y).squeeze()) #shape mxm
    C = -np.ones(shape = (X.shape[0],)) #shape of c -> (m,)
    A = np.eye(N = X.shape[0]) #shape of A -> (m,m)   its an identity matrix
    B = np.zeros(shape = (X.shape[0],)) #shape (m, )
    B = B + c
    E = y.T
    d = np.zeros(shape = (1,))    
    return Q, C, A, B, E, d

def alphastar(Q, c, A, B, E, d): 
    x = cvxopt_solve_qp(Q, c, A, B, E, d)
    x = np.array(x)
    x = x.squeeze()
    return x

def predict(x, X, y, alpha, sigma): #prediction part in lecture note (my note)
    s = []
    for j in range(x.shape[0]): #loop for each points to predict, here x represent test sets
        ss = 0
        for i in range(X.shape[0]):  #loop for each points in training set, X represent training sets
            ss += alpha[i]*y[i]*np.exp((-(X[i]-x[j])@(X[i]-x[j]))/(2*sigma*sigma))
        s.append(ss)
    s = np.array(s)
    s[s >= 0] = 1
    s[s < 0] = -1
    return s

s = [1,10] #RBF parameter, greater the value less curvature
cc = [0.0001, 1]   #higher c -> more training error
for sigma in s:
    for c in cc:
        K_train = gausskernel(X_train, sigma)
        print("K_ shape train", K_train.shape)
        Q_train, C, A, B, E, d = quadparams(X_train, y_train, K_train, c)

        alpha_star_train = cvxopt_solve_qp(Q_train, C, A, B, E, d)

        y_pred = predict(X_train, X_train, y_train, alpha_star_train, sigma)
        accuracy = np.sum(y_train == y_pred)/y_train.size
        print(f"Accuracy is for c={c} and sigma={sigma} is: {accuracy}")

        #plot
        x_series = np.linspace(np.amin(X_train[:,0]),np.amax(X_train[:,0]), 100)
        y_series = np.linspace(np.amin(X_train[:,1]),np.amax(X_train[:,1]), 100)

        x_mesh, y_mesh = np.meshgrid(x_series, y_series)

        x_mesh = x_mesh.reshape(-1, 1)
        y_mesh = y_mesh.reshape(-1, 1)

        mesh = np.append(x_mesh, y_mesh, axis=1)
        y_pred = predict(mesh, X_train, y_train, alpha_star_train, sigma)

        x_mesh = x_mesh.reshape(100, 100)
        y_mesh = y_mesh.reshape(100, 100)
        y_pred = y_pred.reshape(100, 100)

        fig1 = plt.figure(figsize=(8,8))
#         ax = plot_annulus(X1, X2)
        plt.scatter(X_train[:,0], X_train[:,1], c=y_train)
        plt.pcolormesh(x_mesh, y_mesh, y_pred, cmap='viridis', shading='auto', alpha=0.1)
        plt.show()
```
